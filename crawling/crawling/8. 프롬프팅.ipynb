````python
from posix import system

def translate_text_chatgpt(text,src_lang,trg_lang):
    def build_fewshot(src_lang,trg_lang):
        src_examples = example[src_lang]
        trg_examples = example[trg_lang]
        fewshot_messages = []
        # ë‘ ë°”í€´ì”© ëˆë‹¤.
        for src_text, trg_text in zip(src_examples, trg_examples):
            fewshot_messages.append({"role":"user","content":src_text})     # ì˜¤ëŠ˜ ë‚ ì”¨ ì–´ë•Œ?
            fewshot_messages.append({"role":"assistant","content":trg_text})    # How is the weather? | ä»Šæ—¥ã®å¤©æ°—ã¯ã©ã†ï¼Ÿ
        return fewshot_messages
    # // ë‚´ë¶€í•¨ìˆ˜ ê¸°ëŠ¥ ì—¬ê¸°ê¹Œì§€
    system_instruction = f'assistantëŠ” ë²ˆì—­ì•±ìœ¼ë¡œ ë™ì‘í•œë‹¤. {src_lang}ë¥¼ {trg_lang}ë¡œ ì ì ˆí•˜ê²Œ ë²ˆì—­ëœ í…ìŠ¤íŠ¸ë§Œ ì¶œë ¥í•œë‹¤.'
    fewshot_messages = build_fewshot(src_lang,trg_lang)
    # *fewshow_messages ë¥¼ ë”•ì…”ë„ˆë¦¬ë¡œ ê·¸ëŒ€ë¡œ ê½‚ì•„ì£¼ëŠ” ìš©ë„
    messages = [{"role":"system","content":system_instruction}, *fewshot_messages, {"role":"user","content":text}]
    response = client.chat.completions.create(
        model = 'gpt-4.1-nano-2025-04-14',
        messages = messages
    )
    return response.choices[0].message.content
````

````python
def gradio_translate(text,src_lang,trg_lang):
    if not text.strip():
        return "ë²ˆì—­í•  ë‚´ìš©ì„ ì…ë ¥í•˜ì„¸ìš”."
    if src_lang == trg_lang:
        return "ì›ë³¸ ì–¸ì–´ì™€ ë²ˆì—­ ì–¸ì–´ê°€ ê°™ìŠµë‹ˆë‹¤. ë‹¤ë¥¸ ì–¸ì–´ë¥¼ ì„ íƒí•˜ì„¸ìš”."
    return translate_text_chatgpt(text,src_lang,trg_lang)

with gr.Blocks() as demo:
    gr.Markdown("# ì´ˆê°„ë‹¨ ë²ˆì—­ ì•±")
    text_input = gr.Textbox(label='ë²ˆì—­í•  ë‚´ìš©ì„ ì…ë ¥í•˜ì„¸ìš”',placeholder="ì—¬ê¸°ì— ë‚´ìš©ì„ ì…ë ¥í•˜ì„¸ìš”.")
    src_lang_dropdown = gr.Dropdown(choices=['í•œêµ­ì–´','ì˜ì–´','ì¼ë³¸ì–´'], value = "ì˜ì–´", label="ë²ˆì—­í•  ì–¸ì–´ë¥¼ ì„ íƒí•˜ì„¸ìš”.")
    trg_lang_dropdown = gr.Dropdown(choices=['ì˜ì–´','í•œêµ­ì–´','ì¼ë³¸ì–´'], value = 'í•œêµ­ì–´', label = "ë²ˆì—­ë  ì–¸ì–´ë¥¼ ì„ íƒí•˜ì„¸ìš”.")
    translate_button =gr.Button("ë²ˆì—­í•˜ê¸°")
    output_text = gr.Textbox(label='ë²ˆì—­ ê²°ê³¼', placeholder='ë²ˆì—­ëœ ê²°ê³¼ê°€ ì—¬ê¸°ì— í‘œì‹œë©ë‹ˆë‹¤.', interactive=False)

    translate_button.click(
        gradio_translate,
        inputs = [text_input, src_lang_dropdown, trg_lang_dropdown],
        outputs = output_text
    )

demo.launch()
````
